```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
from scipy.stats import norm

%matplotlib inline
#initial conditions
mu = 10
sigma = 2
n= 100
bin_size = 0.5
#this is the data for the Guassian Distribution
data = np.random.normal(mu, sigma, n)
fig, ax = plt.subplots()
#plot the data
y,binEdges, x = plt.hist(data, bins = np.arange(min(data), max(data), bin_size))
bincenters = 0.5*(binEdges[1:]+binEdges[:-1])

#Poisson Distribution and error
error = np.sqrt(n)
plt.errorbar(bincenters, y, yerr= error, fmt="o", color="0", capsize = 2)

#overplot theoretical Guassian Curve
x = np.linspace(min(data), max(data), n)
scale = len(data)
plt.plot(x, norm.pdf(x, mu, sigma)*scale)
yg = (norm.pdf(x, mu, sigma)*scale)[:(np.shape(y)[0])]
np
#Find Reduced Chi Squared %%%%%%%%%% still have to figure this out %%%%%%%%%
print(np.shape(y),np.shape(yg))
z = ((y - yg) / error)
chi2 = np.sum(z ** 2)/(n-1)
#add Chi^2 label
plt.text(0.90, 0.90, r'$\chi^2{\rm 100} = %.2f$' % chi2 , ha = 'right', va = 'top', fontsize = 12, fontweight = 'bold', color='b',
        transform = ax.transAxes)
plt.xlabel('Number')
plt.ylabel('Occurences')
plt.show()



#Now do the same for 1M samples
n2 = 1000000
#generate data
data2 = np.random.normal(mu, sigma, n2)
y2, binEdges2, x2 = plt.hist(data2, bins = np.arange(min(data2), max(data2), bin_size), color = 'r')
bincenters2 = 0.5*(binEdges2[1:]+binEdges2[:-1])
#Guassian Curve
x2 = np.linspace(min(data), max(data), n2)
scale = len(data2)*bin_size
plt.plot(x2, norm.pdf(x2, mu, sigma)*scale)
#error bars
error2 = np.sqrt(n2)
plt.errorbar(bincenters2, y2, yerr= error2, fmt="o", color="0", capsize = 2)
#chi^2   %%%%%%%% not correct f%%%%%%%%%
z2 = ((y2 - mu)/ error2)
chi2_2 = np.sum(z2 ** 2)/(n2-1)
#add labels

plt.text(0.90, 0.80, r'$\chi^2{\rm 1M} = %.2f$' % chi2_2 , ha = 'right', va = 'top', fontsize = 12, fontweight = 'bold', color='r',
        transform = ax.transAxes)
plt.xlabel('Number')
plt.ylabel('Occurences')
plt.show()
```
